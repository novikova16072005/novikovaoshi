```python
def solve_quad(a: float, b: float, c: float):
    discr = b**2 - 4*a*c

    if discr < 0:
        return "No roots"
    elif a == 0:
      return "No roots"
    elif discr == 0:
        root = -b / (2*a)
        return root
    else:
        sqrt_discr = discr ** 0.5
        root1 = (-b + sqrt_discr) / (2*a)
        root2 = (-b - sqrt_discr) / (2*a)
        return root1, root2

print(solve_quad(a=1, b=-2, c=1))
print(solve_quad(a=1, b=0, c=-1))
print(solve_quad(a=0, b=2, c=-1))
```

```python
def merge(a: List[int], b: List[int]) -> List[int]:
    merged = []
    i = j = 0

    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            merged.append(a[i])
            i += 1
        else:
            merged.append(b[j])
            j += 1

    while i < len(a):
        merged.append(a[i])
        i += 1

    while j < len(b):
        merged.append(b[j])
        j += 1

    return merged

print(merge([1, 3, 3, 5], [3, 5, 5, 7]))
```

```python
def MergeSort(L: List[int]) -> List[int]:
    def merge(left: List[int], right: List[int]) -> List[int]:
        merged = []
        i = k = 0

        while i < len(left) and k < len(right):
            if left[i] <= right[k]:
                merged.append(left[i])
                i += 1
            else:
                merged.append(right[k])
                k += 1

        while i < len(left):
            merged.append(left[i])
            i += 1

        while k < len(right):
            merged.append(right[k])
            k += 1

        return merged

    if len(L) <= 1:
        return L

    mid = len(L) // 2
    left = MergeSort(L[:mid])
    right = MergeSort(L[mid:])

    return merge(left, right)


print(MergeSort([3, 1, 4, 3, 7, 5]))
```

```python
def compare(file):
    data = pd.read_csv(file)
    male = (data['Sex'] == 'male').sum()
    female = (data['Sex'] == 'female').sum()

    if male > female:
        return "Чоловіків більше"
    elif male < female:
        return "Жінок більше"
    else:
        return "Чоловіків і жінок порівну"

result = compare('./titanic.csv')
print(result)
```

```python
def passenger_data(file):
    data = pd.read_csv(file)

    male = data[data['Sex'] == 'male']
    female = data[data['Sex'] == 'female']

    avg_age = data['Age'].mean()

    avg_age_male = male['Age'].mean()
    avg_age_female = female['Age'].mean()

    max_age_male = male['Age'].max()
    min_age_male = male['Age'].min()
    max_age_female = female['Age'].max()
    min_age_female = female['Age'].min()

    return avg_age, avg_age_male, avg_age_female, max_age_male, min_age_male, max_age_female, min_age_female

# Приклад використання:
avg_age, avg_age_male, avg_age_female, max_age_male, min_age_male, max_age_female, min_age_female = passenger_data('./titanic.csv')

print(f"Середній вік пасажирів: {avg_age:.2f}")
print(f"Середній вік чоловіків: {avg_age_male:.2f}")
print(f"Середній вік жінок: {avg_age_female:.2f}")
print(f"Максимальний вік серед чоловіків: {max_age_male}")
print(f"Мінімальний вік серед чоловіків: {min_age_male}")
print(f"Максимальний вік серед жінок: {max_age_female}")
print(f"Мінімальний вік серед жінок: {min_age_female}")
```
