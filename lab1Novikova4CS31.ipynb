завдання 11

```python
clear_df = df.copy()
```

```python
launched_med = clear_df["launched"].median()
clear_df["launched"] = clear_df["launched"].fillna(launched_med).astype('int64')
display(clear_df.head(10))
```

завдання 12

```python
clear_df = clear_df.drop(["source", "destroyed_details", "carrier", "not_reach_goal"], axis=1)
display(clear_df.head(5))
```

завдання 13

```python
df_before = len(clear_df)
clear_df = clear_df.drop_duplicates()
df_after_duplicates = len(clear_df)

clear_df['time_start'] = pd.to_datetime(clear_df['time_start'], errors='coerce')
clear_df['time_end'] = pd.to_datetime(clear_df['time_end'], errors='coerce')
clear_df['attack_duration'] = (clear_df['time_end'] - clear_df['time_start']).dt.total_seconds()

clear_df = clear_df[clear_df['attack_duration'] > 0]
df_after_duration = len(clear_df)

clear_df = clear_df.reset_index(drop=True)

rows_removed = df_before - df_after_duration
```

завдання 14

```python
display(clear_df[clear_df.isna().any(axis=1)])
```

завдання 15

```python
percentage = clear_df["destroyed"]/clear_df["launched"]*100

print("The average percentage is: ", percentage.median())
```

завдання 16

```python
top_models = (
    clear_df.groupby("model")["launched"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

top_models.head(10).to_json("top_launched_models.json", orient='records', lines=True)

display(top_models.head(10))
```

завдання 17

```python
clear_df["hits"] = clear_df["launched"] - clear_df["destroyed"]

day_with_most_hits = (
    clear_df.sort_values(by="hits", ascending=False)
    .head(1)
    [["time_start", "hits"]]
)

display(day_with_most_hits)
```

завдання 18

```python
clear_df["attack_duration"] = clear_df["time_end"] - clear_df["time_start"]

clear_df["year"] = clear_df["time_start"].dt.year
clear_df["month"] = clear_df["time_start"].dt.month

attacks_by_months_df = clear_df[["year", "month", "attack_duration"]]

min_and_max_attacks_df = (
    attacks_by_months_df.groupby(["year", "month"])["attack_duration"]
    .agg(['min', 'max'])
    .reset_index()
)

display(min_and_max_attacks_df)

min_and_max_attacks_df.to_csv("min_and_max_attacks_for_each_month.csv", index=False)
```
